```python
import pandas as pd
```

```python
INPUT_FILE = r"../data/drug_consumption_transformed.csv"
OUTPUT_FILE = r"../data/drug_consumption_numeric.csv"
```

```python
df = pd.read_csv(INPUT_FILE)
df
```

```python
COLUMN_ORDERS = {
    "age": ["18-24", "25-34", "35-44", "45-54", "55-64", "65+"],
    "education":  ["Left school", "Some college or university, no certificate or degree", "Professional certificate/diploma",
                    "University degree", "Masters degree", "Doctorate degree"],
}
```

```python
for col, orders in COLUMN_ORDERS.items():
    n_levels = len(orders)
    mapping = dict(zip(orders, range(n_levels)))
    df[col] = df[col].map(mapping)
df
```

```python
df.to_csv(OUTPUT_FILE, index = False)
```
