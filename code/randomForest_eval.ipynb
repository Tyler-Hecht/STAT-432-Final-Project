```python
from TC import TC
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
```

```python
TC_CAT_FILE = r"../data/randomForest/rf_tune_cat.csv"
TC_NUM_FILE = r"../data/randomForest/rf_tune_num.csv"

OUTPUT_CAT_PLOT = r"../plots/rf_TC_cat.png"
OUTPUT_NUM_PLOT = r"../plots/rf_TC_num.png"
```

```python
TC_cat = pd.read_csv(TC_CAT_FILE)
TC_num = pd.read_csv(TC_NUM_FILE)

TC_cat_pivot = TC_cat.pivot(index = "nodesize", columns = "mtry", values = "TC_f_rf")
TC_num_pivot = TC_num.pivot(index = "nodesize", columns = "mtry", values = "TC_rf")
```

```python
TC_num_pivot.min().min()
```

```python
plt.pcolor(TC_cat_pivot, cmap = "viridis", vmin = 2975, vmax = 3900)
plt.xticks(np.arange(len(TC_cat_pivot.columns))+0.5, TC_cat_pivot.columns)
plt.yticks(np.arange(len(TC_cat_pivot.index))+0.5, TC_cat_pivot.index)
plt.ylabel(r"nodesize")
plt.xlabel(r"mtry")
plt.title("Total Misclassification Cost for Random Forest (Categorical)")
plt.colorbar()

plt.savefig(OUTPUT_CAT_PLOT, dpi = 300)
```

```python
plt.pcolor(TC_num_pivot, cmap = "viridis", vmin = 2975, vmax = 3900)
plt.xticks(np.arange(len(TC_num_pivot.columns))+0.5, TC_num_pivot.columns)
plt.yticks(np.arange(len(TC_num_pivot.index))+0.5, TC_num_pivot.index)
plt.ylabel(r"nodesize")
plt.xlabel(r"mtry")
plt.title("Total Misclassification Cost for Random Forest (Numerical)")
plt.colorbar()

plt.savefig(OUTPUT_NUM_PLOT, dpi = 300)
```
